```python
data = []
for p in range(1,50):
    url = f'https://www.daraz.com.bd/bedroom-furniture/?ajax=true&page={p}&spm=a2a0e.home.cate_10_4.1.735212f7A3iq7k'
    req = requests.get(url)
    js = req.json()
    for i in range(40):
        d = {
        'name' : js['mods']['listItems'][i]['name'],
        'nid' : js['mods']['listItems'][i]['nid'],
        'price' : js['mods']['listItems'][i]['price'],
        'priceShow' : js['mods']['listItems'][i]['priceShow'],
        'image' : js['mods']['listItems'][i]['image'],
        'location' : js['mods']['listItems'][i]['location'],
        'brandName' : js['mods']['listItems'][i]['brandName'],
        'sellerId' : js['mods']['listItems'][i]['sellerId'],
        'sellerName' : js['mods']['listItems'][i]['sellerName'],
        'inStock' : js['mods']['listItems'][i]['inStock'],
        }
    print(d)
    data.append(d)
    
```
